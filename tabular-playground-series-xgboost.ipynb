```python
import xgboost
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('data/train.csv')
data.head()
```

```python
le = LabelEncoder()
for col in data.columns:
    if col.__contains__('cat'):
        data[col] = le.fit_transform(data[col])
data.head()
```

```python
Ytrain = data['target']
Xtrain = data.drop(['target', 'id'], axis=1)
```

```python
xgb_r = xgboost.XGBRegressor()
xgb_r.fit(Xtrain, Ytrain)
xgb_r.score(Xtrain, Ytrain)
```

```python
test_data = pd.read_csv('data/test.csv')
test_data.head()
```

```python
for col in test_data.columns:
    if col.__contains__('cat'):
        test_data[col] = le.fit_transform(test_data[col])
test_data.head()
```

```python
test_df = test_data.drop(['id'], axis=1)
prediction = xgb_r.predict(test_df)
test_data['target'] = prediction
test_data.head()
```

```python
for col in test_data.columns:
    if col.__contains__('cat') or col.__contains__('cont'):
        test_data.drop([col], axis=1, inplace=True)
test_data.head()
```

```python
test_data.to_csv('submission_xgb.csv', index=False)
```
